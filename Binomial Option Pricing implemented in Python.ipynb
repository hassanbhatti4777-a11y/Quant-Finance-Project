```python
import math
```

```python
import numpy as np
```

```python
Asset = 100
Volatility = 0.2
IntRate = 0.05
Strike = 100
Expiry = 1
NoSteps = 3
```

```python
import numpy as np

# Initialize S and V as NumPy arrays
S = np.zeros(NoSteps + 1)
V = np.zeros(NoSteps + 1)

time_step = Expiry / NoSteps
DiscountFactor = math.Exp(-IntRate * time_step)
temp1 = math.Exp((IntRate + Volatility * Volatility * time_step))
```

```python
temp2 = 0.5 * (DiscountFactor + temp1)
u = temp2 + math.sqrt(temp2 * temp2 -1)
d = 1 / u
p = (math.exp(IntRate * time_step) - d) / (u - d)
```

```python
S[0] = Asset
for n in range(1, NoSteps + 1):
    for j in range(n, 0, -1):
        S[j] = u * S[j - 1]
    S[0] = d * S[0]
```

```python
def Payoff(stock_price, strike_price):
    return max(0, stock_price - strike_price)

for j in range(NoSteps + 1):
    V[j] = Payoff(S[j], Strike)
```

```python
for n in range(NoSteps - 1, -1, -1):
    for j in range(n):
        V[j] = (p * V[j + 1] + (1 - p) * V[j]) * DiscountFactor
```

```python
Price = V[0]
```

```python
print(f"The option price is: {Price}")
```
